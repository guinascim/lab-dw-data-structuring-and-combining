## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url)
```

```python
new_df = pd.concat([df, df2, df3], axis=0)
new_df
```

```python
new_df.shape
```

```python
new_df.columns = new_df.columns.str.lower()
```

```python
new_df.columns = new_df.columns.str.strip().str.replace(' ', '_')
```

```python
new_df.columns.unique()
```

```python
new_df.isnull().sum()
```

```python
new_df.total_claim_amount.fillna(new_df.total_claim_amount.mean(), inplace=True)
```

```python
new_df.income.fillna(new_df.income.mean(), inplace=True)
```

```python
new_df.replace(to_replace="Femal", value="F", inplace=True)
new_df.replace(to_replace="female", value="F", inplace=True)
new_df.replace(to_replace="Male", value="M", inplace=True)
```

```python
#new_df.gender.fillna("F", inplace=True)
```

```python
new_df.replace(to_replace="AZ", value="Arizona", inplace=True)
new_df.replace(to_replace="Cali", value="California", inplace=True)
new_df.replace(to_replace="WA", value="Washington", inplace=True)
```

```python
new_df.replace(to_replace="Bachelors", value="Bachelor", inplace=True)
```

```python
new_df['customer_lifetime_value'] = new_df['customer_lifetime_value'].str.replace(r'%', '', regex=True)
```

```python
new_df['customer_lifetime_value'].fillna(method='ffill', inplace=True)
new_df['customer_lifetime_value'].fillna(method='bfill', inplace=True)
```

```python
new_df.replace(to_replace="Sports Car", value="Luxury", inplace=True)
new_df.replace(to_replace="Luxury SUV", value="Luxury", inplace=True)
new_df.replace(to_replace="Luxury Car", value="Luxury", inplace=True)
```

```python
new_df = new_df.rename(columns={'st': 'state'})
```

```python
new_df.drop_duplicates(inplace=True)
```

```python
#https://www.statology.org/pandas-merge-columns-with-same-name/

def same_merge(x): return ','.join(x[x.notnull()].astype(str))

df_merged = new_df.groupby(level=0, axis=1).apply(lambda x: x.apply(same_merge, axis=1))
```

```python
df_merged.columns
```

```python
df_merged.isnull().sum()
```

```python
df_merged.isnull().any()
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
